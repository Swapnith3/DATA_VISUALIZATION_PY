### Data visualization

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("titanic-training-data.csv")
```

```python
df.head()
```

```python
df.tail()
```

```python
df=df.drop(["PassengerId","Name","Ticket","Cabin"],axis=1)
df.head()
```

```python
df.head()
```

### Univariate -- Single Varaible

```python
sns.countplot(x="Survived",data=df)   ### we use countplot when variable is Categorical
plt.show()
```

```python
sns.countplot(x="Embarked",data=df)
plt.show()
```

```python
sns.countplot(x="Pclass",data=df)
plt.show()
```

### Boxplot ----- (we use boxplot when the value is Numerical and to fing the outliers as well aa q1,q2,q3 values)

```python
df.describe()
```

```python
sns.boxplot(x="Age",data=df)
plt.show()
```

```python
sns.boxplot(x="SibSp",data=df)
plt.show()
```

```python
sns.boxplot(x="Fare",data=df)
plt.show()
```

```python
sns.boxplot(x="Parch",data=df)
plt.show()
```

### Histogram --- (we use histogram to study the data and shape of data)

```python
plt.hist(x='Age',data=df)
plt.show()
```

```python
plt.hist(x="Parch",data=df)
plt.show()
```

```python
plt.hist(x="SibSp",data=df)
plt.show()
```

### Bivariate --- Two Attributes

```python
sns.countplot(x="Survived",hue="Pclass",data=df)
plt.show()
```

```python
sns.countplot(x="Survived",hue="Sex",data=df)
plt.show()
```

```python
sns.countplot(x="Survived",hue="Embarked",data=df)
plt.show()
```

### relplot -- used for numericals

```python
sns.relplot(x="Age",y="Parch",data=df)
```

```python
df.describe()
```

```python
sns.relplot(x="Parch",y="Fare",data=df)
```

### Catplot -- for one categorical and one numerical

```python
sns.catplot(x="Survived",y="Age",data=df)
```

```python
sns.catplot(x="Sex",y="Age",data=df)
```

```python
sns.catplot(x="Survived",y="SibSp",data=df)
```

```python
sns.catplot(x="SibSp",y="Parch",data=df)
```

```python
sns.catplot(x="Pclass",y="Age",data=df,kind="box")
```

```python
sns.catplot(x="Sex",y="Age",data=df,kind="swarm")
```

```python
sns.catplot(x="Sex",y="Pclass",data=df,kind="violin")
```

```python
sns.pairplot(data=df)    #All Graphs 
```

### Multivariate Analysis

```python
sns.catplot(x="Survived",y="Age",hue="Sex",data=df)
```

```python
sns.catplot(x="Survived",y="SibSp",hue="Parch",data=df)
```

```python
sns.catplot(x="Survived",y="Age",hue="Embarked",data=df)
```

```python
sns.boxplot(x="Survived",y="Fare",hue="Sex",data=df)
```
